```python
import numpy as np

A = np.array([
    [6,  -4,  -8],
    [-3, 10,  11],
    [2,  -4,  -4]
], dtype=float)

A6 = A / 6.0

# I Use a high power
k = 100
B = np.linalg.matrix_power(A6, k)
B_rounded = np.round(B, 6)

print(f"(A/6)^{k} ≈")
print(B_rounded)

# I will take the columns
c1 = B[:, 0]
c2 = B[:, 1]
c3 = B[:, 2]

print("\nFirst column c1:", c1)
print("Second column c2:", c2)
print("Third column c3:", c3)

# Here, I show ratios to see if they are multiples of each other
print("\nRatios c2 / c1:", c2 / c1)
print("Ratios c3 / c1:", c3 / c1)
```
