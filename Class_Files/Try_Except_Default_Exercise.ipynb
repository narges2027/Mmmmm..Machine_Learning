### Basic Sum and Product

```python
def new_sum(a,b):
    return int(a+b-1)
```

```python
def new_prod(a,b):
    return int(a+b-a*b)
```

### Revised Sum

```python
def new_sum2(a=1,b=1):
    s = 1
    try:
        if type(a) != int or type(b) != int:
            raise TypeError
        else:
            s = int(a+b-1)
    except TypeError:
        print("Your inputs are the wrong data type, a=%s is %s and b=%s is %s.\n"%(str(a),type(a),str(b),type(b)))
        s = TypeError
    except Exception as Err:
        print("There was an error, check your inputs.\n",Err)
        s= Err
    return s
```
