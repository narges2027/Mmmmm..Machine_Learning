### Other Useful Things

```python
# This is a for loop
B=A
for i in range(5):
    print(B)
    B = np.matmul(B,A)
```

```python
# While Loop
B = A
i = 0
while i<5:
    print(B)
    B = np.matmul(B,A)
    i += 1
```
